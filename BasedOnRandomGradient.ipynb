```python
#Imports
import math
import random
import matplotlib.pyplot as plt

#Implemented same code from above into a class
class newGradient:
  def __init__(self, xCent, yCent, radius, p):
    self.gradientField = []
    alphaGradient = []
    xG = []
    yG = []
    for y in range (0,20):
      for x in range (0,20):
        xG.append(x)
        yG.append(y)
        distanceFromG1 = math.sqrt((x-9.5)**2+(y-9.5)**2)
        distanceFromG2 = math.sqrt((x-xCent)**2+(y-yCent)**2)
        if(distanceFromG1<=5 and distanceFromG2<=radius):
          if(1/(distanceFromG1**p+0.01) + 1/(distanceFromG2**p+0.01)>=1):
            alphaGradient.append(1)
          else:
            alphaGradient.append(1/(distanceFromG1**p+0.01) + 1/(distanceFromG2**p+0.01))
        elif(distanceFromG1<=5):
          if((1/(distanceFromG1**p+0.01))>=1):
            alphaGradient.append(1)
          else:
            alphaGradient.append(1/(distanceFromG1**p+0.01))
        elif(distanceFromG2<=radius):
          if(1/(distanceFromG2**p+0.01)>=1):
            alphaGradient.append(1)
          else:
            alphaGradient.append(1/(distanceFromG2**p+0.01))
        else:
          alphaGradient.append(0)
    self.gradientField = [xG,yG,alphaGradient]
  def alphaVals(self):
    return self.gradientField


#Generate random gradient, can help test the ability of the model to adapt
steepness = 1
def randomGradient():
  tempGradient = newGradient(random.randint(0,20), random.randint(0,20), random.randint(4,8), steepness)
  graphGradient = tempGradient.alphaVals()
  return graphGradient
```

```python
g = randomGradient()
plt.scatter(g[0],g[1], alpha = g[2], color = 'green', s = 100)
```
