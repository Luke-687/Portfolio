```python
#Consider a 20x20 gride with a point of green at 10,10, make a gradient fill out from every point up to 6 away from 10,10:
import math
import matplotlib.pyplot as plt

XYColorAlpha = []
for j in range(0,20):
  for i in range(0,20):
    distanceFromSource = math.sqrt((j-10)**2+(i-10)**2)
    if(distanceFromSource<=7 and not (distanceFromSource==0)):
      XYColorAlpha.append([i,j,"green", 1/distanceFromSource,10/distanceFromSource])
    elif(distanceFromSource==0):
      XYColorAlpha.append([i,j,"green", 1,10])
    else:
      XYColorAlpha.append([i,j,"green", 0, -10])
x=[]
y=[]
color=[]
alphaG=[]
rewardG=[]
for item in XYColorAlpha:
  x.append(item[0])
  y.append(item[1])
  color.append(item[2])
  alphaG.append(item[3])
  rewardG.append(item[4])

plt.scatter(x,y,c=color, alpha=alphaG)
```
