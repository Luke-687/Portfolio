```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
class tipShape:
  def __init__(self):
    self.xVal = []
    self.yVal = []
  def defineShape(self):
    currentX = 4.5
    for x in range (0,101):
      currentX = 4.5 + 0.045*x
      yTop = math.sqrt(currentX-4.5) + 4.5
      yBottom = -1 * math.sqrt(currentX-4.5) + 4.5
      for y in np.arange(yBottom, yTop, 0.01):
        self.xVal.append(currentX)
        self.yVal.append(y)
```

```python
class particleSim:
  def __init__(self, numParticles):
    self.numParts = numParticles
  def movement(self):
    testY = 0
    yStart = []
    change = 9/(self.numParts-1)
    while(testY<=9):
      yStart.append(testY)
      testY+=change
    data = []
    xPos = []
    yPos = []
    for i in range(0, len(yStart)):
      xPos = []
      yPos = []
      y = yStart[i]
      x = 0
      xPos.append(x)
      yPos.append(y)
      while(x<=9):
        if (x + 0.1 - 4.5) >= 0:
            if(not(-1*math.sqrt(x+0.1-4.5)+4.5 <=y<= math.sqrt(x+0.1-4.5) + 4.5)):
              x+=0.1
            elif(abs(y-(math.sqrt(x+0.1-4.5) + 4.5)) > abs(y-(-1*math.sqrt(x+0.1-4.5)+4.5))):
              yInit = y
              y = (math.sqrt(x+0.1-4.5) + 4.5+0.2)
              x+=0.1
            else:
              yInit = y
              y = (-1*math.sqrt(x+0.1-4.5) + 4.5-0.2)
              x+=0.1
        else:
            x += 0.1
        xPos.append(x)
        yPos.append(y)
      data.append([xPos,yPos])
    return data
```

```python
tipTest = tipShape()
tipTest.defineShape()

particleTest = particleSim(100)
graphingData = particleTest.movement()

for i in range(0, 60):
  plt.xlim(0,9)
  plt.ylim(0,9)
  plt.xticks(range(10))
  plt.yticks(range(10))
  pointsX = []
  pointsY = []
  for p in range(0, len(graphingData)):
    xPoint = graphingData[p][0][i]
    yPoint = graphingData[p][1][i]
    pointsX.append(xPoint)
    pointsY.append(yPoint)
  plt.scatter(tipTest.xVal, tipTest.yVal, color = 'pink', alpha = 0.1)
  plt.scatter(pointsX, pointsY, s = 100)
  plt.show()
```
