```python
import math
import numpy as np

class tipShape:
  def __init__(self):
    self.xVal = []
    self.yVal = []
  def defineShape(self):
    currentX = 4.5
    for x in range (0,101):
      currentX = 4.5 + 0.045*x
      yTop = math.sqrt(currentX-4.5) + 4.5
      yBottom = -1 * math.sqrt(currentX-4.5) + 4.5
      for y in np.arange(yBottom, yTop, 0.01):
        self.xVal.append(currentX)
        self.yVal.append(y)

test = tipShape()
test.defineShape()

import matplotlib.pyplot as plt
plt.xlim(0,9)
plt.ylim(0,9)
plt.scatter(test.xVal, test.yVal, color = 'pink')
```
