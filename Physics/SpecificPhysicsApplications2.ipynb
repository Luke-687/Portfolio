```python
#consider the vector (x,y,z) for initial velocities, and then (ax,ay,az) for the acceleration on each vector (az=-10)
#calculate their position after t seconds:

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#set up vectors + time
def finalPositions(velocities, acceleration, tTot):

  acceleration.append(-10) # az = -10

  #Always begin at origin

  x=[0]
  y=[0]
  z=[0]

  for i in range(1, tTot):
    t = i
    #for x:
    changeX = velocities[0]*t + 0.5*(acceleration[0])*(t*t)
    #for y:
    changeY = velocities[1]*t + 0.5*(acceleration[1])*(t*t)
    #for z:
    changeZ = 0
    changeZ = velocities[2]*t+1/2*(-10)*t*t
    if(changeZ<=0):
      changeZ = 0

    x.append(changeX)
    y.append(changeY)
    z.append(changeZ)


  return x,y,z
```

```python
#Enter, velocities, accelerationX/Y, time
x,y,z = finalPositions([10,0,15],[2,5], 10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

if(max(y)>max(x)):
  x.append(max(y))
  y.append(0)
  z.append(0)
else:
  x.append(0)
  y.append(max(x))
  z.append(0)
print(x)
print(y)
print(z)

ax.scatter(x, y, z)
plt.show()
```
