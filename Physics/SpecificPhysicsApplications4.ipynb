```python
#Essentially just a dynamics question of something on an inclined plane
#provide angle, mass, and coeff. friction, find the acceleration, and with that acceleration is found, and then graphed out, showing acceleration vs velocity vs position
import math

angleTheta = math.pi/4
mass = 10
friction = 0.15
acceleration = 0
g=10

acceleration = (mass*g*math.sin(angleTheta) - mass*g*math.cos(angleTheta)*friction)/mass

#Graph acc vs velocty vs position for first 10 seconds
import matplotlib.pyplot as plt

xA=[]
yA=[]
xV=[]
yV=[]
xP=[]
yP=[]

for i in range(0,10):
  xA.append(i)
  yA.append(acceleration)
  xV.append(i)
  yV.append(i*acceleration)
  xP.append(i)
  currentPos = 1/2 * (acceleration) * (i**2)
  yP.append(currentPos)


plt.plot(xA, yA, color = "black", label = "acceleration")
plt.plot(xV, yV, color="green", label="velocity")
plt.plot(xP,yP, color="blue", label="position")
plt.legend()
```
