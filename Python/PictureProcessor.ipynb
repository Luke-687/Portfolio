Now that the `pillow-heif` library is installed, I'll modify your code to use it to open the HEIC file.

```python
from PIL import Image
import numpy as np
import pillow_heif

# Register the HEIF opener with Pillow
pillow_heif.register_heif_opener()

img = Image.open("/content/IMG_0820.HEIC")
pixelData = np.array(img)

# There was a typo in the original code, changing 'pixleData' to 'pixelData'
print(pixelData.shape)
print(pixelData[0,0])
remakeImage = []
for i in range (0,3024):
  for j in range (0, 4032):
    if(pixelData[i,j][0]<100 and pixelData[i,j][1]<100 and pixelData[i,j][2]<100):
      remakeImage.append([j,i,"x"])
    else:
      remakeImage.append([j,i," "])
```

```python
xGraph =[]
yGraph = []
for data_point in remakeImage:
  if(data_point[2] == "x"):
    xGraph.append(data_point[0])
    yGraph.append(data_point[1])

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.scatter(xGraph, yGraph, c="black")
```
