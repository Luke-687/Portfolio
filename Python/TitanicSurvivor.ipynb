```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

dftrain = pd.read_csv("train.csv")
dftrain = dftrain.drop(columns = ["Name", "Ticket", "Embarked", "Cabin"])
dftrain = dftrain.replace("male", 1)
dftrain = dftrain.replace("female", 0)

dftest = pd.read_csv("test.csv")
dftest = dftest.drop(columns = ["Name", "Ticket", "Embarked", "Cabin"])
dftest = dftest.replace("male", 1)
dftest = dftest.replace("female", 0)

X = dftrain.drop(columns = "Survived")
y = dftrain["Survived"]

survivalModel = DecisionTreeClassifier()
survivalModel.fit(X, y)

resultArr = []
for index, row in dftest.iterrows():
  predictions = survivalModel.predict([row])
  resultArr.append(predictions[0])

dfresults = pd.DataFrame()
dfresults["PassengerId"] = dftest["PassengerId"]
dfresults["Survived"] = resultArr
```
