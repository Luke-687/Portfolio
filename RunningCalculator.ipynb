```python
def input_week(d1,d2,d3,d4,d5,d6,d7, df):
  new_row = pd.DataFrame({"D1":[d1], "D2":[d2], "D3":[d3], "D4":[d4], "D5":[d5], "D6":[d6], "D7":[d7], "T":[d1+d2+d3+d4+d5+d6+d7]})
  df = pd.concat([df, new_row], ignore_index=True)
  return df
```

```python
def fix_week(week, day, value, df):
  day = "D" + str(day)
  week = week
  df[day][week] = value
  calc_total(week, df)
```

```python
def clear_week(week, df):
  for i in range (1,7):
    fix_week(week, i,0,df)
  calc_total(week, df)
```

```python
def calc_total(week, df):
  total = 0
  for i in range (1,7):
    total+= df[("D" + str(i))][week]
  df["T"][week] = total
```
