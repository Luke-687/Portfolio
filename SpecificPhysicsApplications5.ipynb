```python
#General idea:
#Simple projectile visualization
import math
import matplotlib.pyplot as plt

#required variables
angleTheta = 0
initVel = float(input("What is the initial velocity (m/s)?"))
angleTheta = (math.pi/180)*float(input("What is the angle (degrees)?"))

time = 0
x=[]
y=[]
t = []
finished = False
while(not finished):
  currentX = time*(initVel*math.cos(angleTheta))
  currentY = initVel*math.sin(angleTheta)*time+(1/2)*(-10)*(time**2)
  if(time>0 and currentY<0):
    finished = True
  else:
    t.append(time)
    y.append(currentY)
    x.append(currentX)
    time+=0.1

#Graph the results of a xPos vs yPos based on variable t, essentially made parametric equation
plt.plot(t,x, color="black", label = "X value vs Time")
plt.plot(t,y, color="blue", label = "Y value vs Time")
plt.xlabel("Time")
plt.ylabel("X or Y")
plt.legend()
```
