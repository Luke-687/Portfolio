```python
#Plan for this program:
#Create a machine learning model based on large amounts of data from a single stock
#Using this single stock, chosen for the sake of consistency and possible patern recognition
#Meant to train a linear regression or decision tree model in order to predict either percent likelihood of stock going up in next day, or whether to invest (yes/no)
#Now go...

import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier

#Split stock data up into 12 hour segments, each index representing 12 years prior to the subsequent index -- allow for paterns to be recognized with time
dfstockdata = pd.read_csv("stocks.csv")
dfstockdata.loc[len(dfstockdata)] = {"Day" : -90.0, "Price": 93.002913715101815}

#Each row includes 16 values, the current stock price, the price over3 the past week (going to skip last and first 14 values as there will be no dat to work off of), and the price in 1 week
FP1W = []
CP = []
CP05 = []
CP1 = []
CP15 = []
CP2 = []
CP25 = []
CP3 = []
CP35 = []
CP4 = []
CP45 = []
CP5 = []
CP55 = []
CP6 = []
CP65 = []
CP7 = []
CP75 = []
CP8 = []
CP85 = []
CP9 = []
CP95 = []
CP10 = []
CP105 = []
CP11 = []
CP115 = []
CP12 = []
CP125 = []
CP13 = []
CP135 = []
CP14 = []

for index, row in dfstockdata.iterrows():
  if(index<=(180-28-14)):
    FP1W.append(row["Price"])
  if(index>=14 and index<=(180-28)):
    CP.append(row["Price"])
  if(index>=15 and index<=(180-27)):
    CP05.append(row["Price"])
  if(index>=16 and index<=(180-26)):
    CP1.append(row["Price"])
  if(index>=17 and index<=(180-25)):
    CP15.append(row["Price"])
  if(index>=18 and index<=(180-24)):
    CP2.append(row["Price"])
  if(index>=19 and index<=(180-23)):
    CP25.append(row["Price"])
  if(index>=20 and index<=(180-22)):
    CP3.append(row["Price"])
  if(index>=21 and index<=(180-21)):
    CP35.append(row["Price"])
  if(index>=22 and index<=(180-20)):
    CP4.append(row["Price"])
  if(index>=23 and index<=(180-19)):
    CP45.append(row["Price"])
  if(index>=24 and index<=(180-18)):
    CP5.append(row["Price"])
  if(index>=25 and index<=(180-17)):
    CP55.append(row["Price"])
  if(index>=26 and index<=(180-16)):
    CP6.append(row["Price"])
  if(index>=27 and index<=(180-15)):
    CP65.append(row["Price"])
  if(index>=28 and index<=(180-14)):
    CP7.append(row["Price"])
  if(index>=29 and index<=(180-13)):
    CP75.append(row["Price"])
  if(index>=30 and index<=(180-12)):
    CP8.append(row["Price"])
  if(index>=31 and index<=(180-11)):
    CP85.append(row["Price"])
  if(index>=32 and index<=(180-10)):
    CP9.append(row["Price"])
  if(index>=33 and index<=(180-9)):
    CP95.append(row["Price"])
  if(index>=34 and index<=(180-8)):
    CP10.append(row["Price"])
  if(index>=35 and index<=(180-7)):
    CP105.append(row["Price"])
  if(index>=36 and index<=(180-6)):
    CP11.append(row["Price"])
  if(index>=37 and index<=(180-5)):
    CP115.append(row["Price"])
  if(index>=38 and index<=(180-4)):
    CP12.append(row["Price"])
  if(index>=39 and index<=(180-3)):
    CP125.append(row["Price"])
  if(index>=40 and index<=(180-2)):
    CP13.append(row["Price"])
  if(index>=41 and index<=(180-1)):
    CP135.append(row["Price"])
  if(index>=42 and index<=(181)):
    CP14.append(row["Price"])




dfstockssplit = pd.DataFrame()
dfstockssplit["CP"] = CP
dfstockssplit["CP-0.5"] = CP05
dfstockssplit["CP-1"] = CP1
dfstockssplit["CP-1.5"] = CP15
dfstockssplit["CP-2"] = CP2
dfstockssplit["CP-2.5"] = CP25
dfstockssplit["CP-3"] = CP3
dfstockssplit["CP-3.5"] = CP35
dfstockssplit["CP-4"] = CP4
dfstockssplit["CP-4.5"] = CP45
dfstockssplit["CP-5"] = CP5
dfstockssplit["CP-5.5"] = CP55
dfstockssplit["CP-6"] = CP6
dfstockssplit["CP-6.5"] = CP65
dfstockssplit["CP-7"] = CP7
dfstockssplit["CP-7.5"] = CP75
dfstockssplit["CP-8"] = CP8
dfstockssplit["CP-8.5"] = CP85
dfstockssplit["CP-9"] = CP9
dfstockssplit["CP-9.5"] = CP95
dfstockssplit["CP-10"] = CP10
dfstockssplit["CP-10.5"] = CP105
dfstockssplit["CP-11"] = CP11
dfstockssplit["CP-11.5"] = CP115
dfstockssplit["CP-12"] = CP12
dfstockssplit["CP-12.5"] = CP125
dfstockssplit["CP-13"] = CP13
dfstockssplit["CP-13.5"] = CP135
dfstockssplit["CP-14"] = CP14
#Should Buy in 0's for no and 1's for yes
dfstockssplit["FP1W"] = FP1W
SB = []
for index, row in dfstockssplit.iterrows():
  if(row["FP1W"] > row["CP"]):
    SB.append(1)
  else:
    SB.append(0)
dfstockssplit["SB"] = SB

#Train stock model based on data
X = dfstockssplit.drop(columns = ["SB", "FP1W"])
y = dfstockssplit["SB"]
stockModel = DecisionTreeClassifier()
stockModel.fit(X,y)

#User input for model
user = []
x = 0
while(x<29):
  if(x==0):
    user.append(input("Enter Price For Day " + str(x/2) + ": "))
  else:
    user.append(input("Enter Price For Day -" + str(x/2) + ": "))
  x = x+1
user = [int(num) for num in user]
prediction = stockModel.predict(pd.DataFrame([user], columns=X.columns))
if(prediction[0] == 1):
  print("\nBuy now!")
else:
  print("\nDon't Buy Please!")
```
